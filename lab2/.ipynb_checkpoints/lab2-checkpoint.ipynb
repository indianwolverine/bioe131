```python
from Bio import SeqIO
import numpy as np

records = list(SeqIO.parse("../seqs.fa", "fasta"))

records_of_interest = ["cy.3", "pi.3", "hu.31", "hu.44", "rh.36"]

for rec in records: 
    if rec.id in records_of_interest:
        print(rec.id, rec.seq)
```

All the reads seem to be from Adeno-associated Virus isolates according to BLAST results from https://blast.ncbi.nlm.nih.gov/Blast.cgi

For example, this is a representative match => "Non-human primate Adeno-associated virus isolate AAVrh.36 capsid protein (VP1) gene, complete cds"

We might not trust some of the annotation BLAST gives us because our sequences are pretty short and could map to unrelated parts of the genome as well.

```python
# this function plots the gc and at percentages of the cluster passed in as an argument
def plot_gc_cluster(cluster):
    records = list(SeqIO.parse("./seqs.aligned.fa", "fasta"))
    records = list(filter(lambda rec: rec.id in cluster, records))

    gc = [0] * len(records[0].seq)

    for index in range(len(records[0].seq)):
        for j in range(len(records)):
            if records[j].seq[index] == "G" or records[0].seq[index] == "C":
                gc[index] += 1

    gc = list(map(lambda rec: rec / len(records), gc))
    at = list(map(lambda rec: 1.0 - rec, gc))

    f, ax = plt.subplots(figsize=(50,25))

    p1 = plt.bar(height=gc, x=list(range(len(gc))), label = 'GC Content')
    p2 = plt.bar(height=at, x=list(range(len(at))), bottom=np.array(gc), label = 'AT Content')

    plt.xlabel("Position in Sequence")
    plt.ylabel("%AT or %GC")

    plt.show()
```

```python
plot_gc_cluster(["hu.31", "hu.32", "hu.14"])
```

```python
plot_gc_cluster(["hu.43", "hu.44", "hu.46", "hu.48"])
```

```python
plot_gc_cluster(["pi.1", "pi.2", "pi.3"])
```

```python
plot_gc_cluster(["cy.3", "cy.4", "cy.5", "cy.6", "rh.13"])
```

There is little variation in gc and at percentages within the clusters since they are all very similar to each other in the first place.

```python
# This function finds the lengths of each sequence in a cluster
def find_lengths(cluster):
    records = list(SeqIO.parse("./seqs.aligned.fa", "fasta"))
    records = list(filter(lambda rec: rec.id in cluster, records))

    lengths = [0] * len(records)

    for index in range(len(records[0].seq)):
        for j in range(len(records)):
            if not (records[j].seq[index] == "-"):
                lengths[j] += 1
                
    return lengths
```

```python
hu_lengths = find_lengths(["hu.31", "hu.32", "hu.14"])
hu2_lengths = find_lengths(["hu.43", "hu.44", "hu.46", "hu.48"])
pi_lengths = find_lengths(["pi.1", "pi.2", "pi.3"])
cy_lengths = find_lengths(["cy.3", "cy.4", "cy.5", "cy.6", "rh.13"])
```

```python
plt.boxplot([hu_lengths, hu2_lengths, pi_lengths, cy_lengths], labels=["hu cluster", "2nd hu cluster", "pi cluster", "cy cluster"])
plt.show()
```

We can see from this box plot that the records within a cluster are mostly all the same length. Only the 2nd hu cluster consisting of hu.43, hu.44, hu.46, and hu.48 shows some deviation from this.
